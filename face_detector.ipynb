### Face Detector with Image

```python
import numpy as np
import argparse
import cv2
import os
from matplotlib import pyplot as plt
from matplotlib import figure
```

```python
prototxt = "deep-learning-face-detection/deploy.prototxt.txt"
model = "deep-learning-face-detection/res10_300x300_ssd_iter_140000.caffemodel"
image = "deep-learning-face-detection/friends.JPG"
confidence = 0.5
```

###### Start to load image

```python
net = cv2.dnn.readNetFromCaffe(prototxt, model) #load model from disk
image = cv2.imread(image) #load input image
```

```python
(height, weight) = image.shape[:2]
    
#pre-processing which includes setting the blob dimensions and normalization with deep neural network.
blob = cv2.dnn.blobFromImage(cv2.resize(image, (300, 300)), 1.0,
                                 (300, 300), (104.0, 177.0, 123.0))
```

```python
net.setInput(blob) #pass the blob by the net to detect faces
detection = net.forward()
```

###### Loop over the detected faces

```python
for i in range(0, detection.shape[2]):
    conf = detection[0, 0, i, 2]
    if conf > confidence:
        # compute the (x, y)-coordinates of the bounding box for the object
        box = detection[0, 0, i, 3:7] * np.array([weight, height, weight, height]) 
        (x_start, y_start, x_end, y_end) = box.astype("int")
        
        # draw the bounding box of the face along with the associated probability
        text = "{:.2f}%".format(conf * 100)
        y = y_start - 10 if y_start - 10 > 10 else y_start + 10
        cv2.rectangle(image, (x_start, y_start), (x_end, y_end), (0, 0, 255), 2)
        cv2.putText(image, text, (x_start, y), cv2.FONT_HERSHEY_SIMPLEX, 0.45, (0, 0, 255), 2)
```

```python
plt.figure(num=None, figsize=(10, 15), dpi=80, facecolor='w', edgecolor='k')
plt.imshow(image)
```

### Face Detector with Video Stream

```python
from imutils.video import VideoStream
import imutils
import time
```

###### Start webcam's video feed

```python
video_stream = VideoStream(src=0).start()
```

```python
time.sleep(2.0)
```

###### Loop over the frames from the video stream

```python
while True:
    frame = video_stream.read() #read frames.
    frame = imutils.resize(frame, width=400) #resize the width of pixels to 400.
 
    (height, width) = frame.shape[:2] # resize the width and height and create blob by using deep neural network.
    blob = cv2.dnn.blobFromImage(cv2.resize(frame, (300, 300)), 1.0,
        (300, 300), (104.0, 177.0, 123.0))
 
    net.setInput(blob)
    im_detections = net.forward() # get the detections and predictions
    
    # loop over the detections
    for i in range(0, im_detections.shape[2]):
        conf_2 = im_detections[0, 0, i, 2]
        if conf_2 < confidence: # eliminate the weak detection by comparing confidence of detected faces.
            continue

        box = im_detections[0, 0, i, 3:7] * np.array([width, height, width, height]) # computethe (x,y) coordinates of bounded box.
        (x_start, y_start, x_end, y_end) = box.astype("int")
         
        #draw bounded box with associated probability    
        text = "{:.2f}%".format(conf_2 * 100)
        y = y_start - 10 if y_start - 10 > 10 else y_start + 10
        cv2.rectangle(frame, (x_start, y_start), (x_end, y_end),
            (0, 0, 255), 2)
        cv2.putText(frame, text, (x_start, y),
            cv2.FONT_HERSHEY_SIMPLEX, 0.45, (0, 0, 255), 2)
        
    #show the output
    cv2.imshow("Frame", frame)
    key = cv2.waitKey(1) & 0xFF
 
    # to break loop, press 'w' key
    if key == ord("w"):
        break
```

```python
cv2.destroyAllWindows()
video_stream.stop()
```
